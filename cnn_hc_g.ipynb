```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, AveragePooling2D
from tensorflow.keras.layers import Activation, BatchNormalization, Dropout, Flatten, Rescaling
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras import optimizers, callbacks
import time, json, os, shutil, argparse
import warnings
warnings.simplefilter("ignore")
```

```python
epochs = 100
batch_size = 160
labels = ["0", "1"]
dir_models = "./Models/HC/G"
dir_training = "./Data/HC/G/"
dir_validation = "./Data/HC/O/Validation"
img_width, img_height, img_depth = 320, 240, 3
path_model = os.path.join(dir_models, 'cnn.h5')
path_history = os.path.join(dir_models, 'cnn.csv')
```

```python
model = Sequential()

k_reg = tf.keras.regularizers.l2(0.0001)
k_init = tf.keras.initializers.RandomNormal(stddev=0.02)

model.add(Rescaling(1./127.5, offset=-1, input_shape=(img_width, img_height, img_depth)))
model.add(MaxPooling2D(pool_size=(4,4)))

model.add(Conv2D(128, (3, 3), kernel_initializer=k_init, kernel_regularizer=k_reg))
model.add(MaxPooling2D())
model.add(Activation('relu'))

model.add(Conv2D(256, (3, 3), kernel_initializer=k_init, kernel_regularizer=k_reg))
model.add(MaxPooling2D())
model.add(Activation('relu'))

model.add(Flatten())
model.add(Dropout(0.5))

model.add(Dense(512, kernel_regularizer=k_reg))
model.add(Activation('relu'))

model.add(Dense(1))
model.add(Activation('sigmoid'))

#model.summary()

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.Nadam(),
              metrics=['accuracy'])
```

```python
train_datagen = ImageDataGenerator()
test_datagen = ImageDataGenerator()

train_generator = train_datagen.flow_from_directory(
    dir_training,
    target_size=(img_width, img_height),
    color_mode="rgb",
    batch_size=batch_size,
    class_mode='binary',
    shuffle=True,
    seed=42)

validation_generator = test_datagen.flow_from_directory(
    dir_validation,
    target_size=(img_width, img_height),
    color_mode="rgb",
    batch_size=batch_size,
    class_mode='binary',
    shuffle=True,
    seed=42)
```

```python
callback_log = callbacks.CSVLogger(path_history, append=True)
callback_model_save = callbacks.ModelCheckpoint(filepath=path_model, save_weights_only=False, monitor='val_accuracy', mode='max', save_best_only=True)
```

```python
time_in = time.time()

hist = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator,
    callbacks=[callback_log, callback_model_save])

time_out = time.time()
print('\n', 'Time cost:', '\n', time_out-time_in)
```
